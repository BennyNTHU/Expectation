```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import floor, ceil
from matplotlib.font_manager import FontProperties
```

假設一團有$m$人,買$n$個商品,求中$k$組之期望值

```python
def expectation(m): 
    sum = 0
    for k in range(1, m+1):
        p = (m-k+1)/m
        sum = sum + 1/p
    return sum
```

```python
vfunc = np.vectorize(expectation, otypes=[float])
```

繪製長條圖

```python
plt.rcParams['figure.figsize'] = (18.0,12.0) 
fig = plt.figure()
ax1 = fig.add_subplot(111)

x = np.arange(1,60,1)
y = vfunc(x)
barlist=ax1.bar(x, y, alpha=.7, color='g')
barlist[9-1].set_color('r')
barlist[11-1].set_color('r')
barlist[13-1].set_color('r')
barlist[52-1].set_color('r')
ax1.set_ylabel("Quantities needs to buy", fontsize='15')
ax1.set_xlabel("Number of group members",fontsize='15')  

plt.yticks(fontsize=15)
for (i,j) in zip(x,y):
    plt.text(i, j, str(ceil(j)), ha='center', va='bottom', fontsize=10, rotation=0)

plt.show()
```
