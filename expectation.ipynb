```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import floor, ceil
from matplotlib.font_manager import FontProperties

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
```

```python
def expectation(m): #一團有m人,買n個商品,中k組之期望值
    sum = 0
    for k in range(1, m+1):
        p = (m-k+1)/m
        sum = sum + 1/p
    return sum

vfunc = np.vectorize(expectation, otypes=[float])
```

```python
x = np.arange(1,33,1)
y = vfunc(x)
```

```python
plt.rcParams['figure.figsize'] = (18.0,12.0) 
fig = plt.figure()
ax1 = fig.add_subplot(111)
barlist=ax1.bar(x, y, alpha=.7, color='g')
barlist[4].set_color('r')
barlist[8].set_color('r')
barlist[11].set_color('r')
barlist[12].set_color('r')
barlist[31].set_color('r')
ax1.set_ylabel("所需買的數量", fontsize='15')
ax1.set_xlabel("團體人數",fontsize='15')  
plt.yticks(fontsize=15)
for (i,j) in zip(x,y):
    plt.text(i, j, str(ceil(j)), ha='center', va='bottom', fontsize=20, rotation=0)
plt.show()
```
